```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"aggregated_data\genres_followers\genres_per_followers.csv")
```

```python
df['quartile_group'] = pd.qcut(df['num_followers'], q=[0, 0.25, 0.5, 0.75, 1], labels=['Q1', 'Q2', 'Q3', 'Q4'])
df['num_followers'] = np.log(df['num_followers'])
```

```python
sns.scatterplot(data=df,x='genre_count',y='num_followers',hue='quartile_group')
```

```python
df.drop(['quartile_group'],axis=1).corr()
```
