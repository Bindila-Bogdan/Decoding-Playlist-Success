```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python

colors = ["#1ED760", "#120D1A", "#82E6A6"]

path = 'PLAYLISTCOMPOSITION/'  
all_files = glob.glob(path + "*.csv")

composition_df = pd.concat((pd.read_csv(f) for f in all_files))

print("Basic Statistics for Playlist Composition:")
print(composition_df.describe())

fig, axs = plt.subplots(3, 1, figsize=(10, 12))
axs[0].hist(composition_df['num_artists'], bins=50, color=colors[0])
axs[1].hist(composition_df['num_albums'], bins=50, color=colors[0])
axs[2].hist(composition_df['num_tracks'], bins=50, color=colors[0])

axs[0].set_title('Number of Artists')
axs[1].set_title('Number of Albums')
axs[2].set_title('Number of Tracks')

for ax in axs:
    ax.set_xlabel('Count')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python

colors = ["#1ED760", "#120D1A", "#82E6A6"]

sns.set()

path = 'PLAYLISTDURATION/' 
all_files = glob.glob(path + "*.csv")

duration_df = pd.concat((pd.read_csv(f) for f in all_files))

# Convert duration from milliseconds to minutes
duration_df['duration_min'] = duration_df['duration_ms'] / 60000
statistics = duration_df['duration_min'].describe().drop('count')

print("Basic Statistics for Playlist Duration:")
print(statistics)

plt.figure(figsize=(10, 6))
sns.histplot(duration_df, x='duration_min', bins=50, color=colors[0])
plt.xlabel('Duration in Minutes')
plt.ylabel('Number of Playlists')
plt.title('Distribution of Playlist Durations')
plt.show()
```
