MOST POPULAR GENRES

The following bar plot shows the frequency of each genre.

```python
import pandas as pd
import matplotlib.pyplot as plt

genre_counts_df = pd.read_csv('data/GENRECOUNTS/genre_counts.csv')

genre_counts_df = genre_counts_df.sort_values(by='count', ascending=False)
plt.figure(figsize=(12, 8))
plt.bar(genre_counts_df['genre'], genre_counts_df['count'])
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.title('Most Popular Genres')
plt.show()
```

This shows the top 20 most popular genres.

```python
genre_counts = pd.read_csv("data/GENRECOUNTS/genre_counts.csv")


# Aggregate the genre counts
genre_counts_aggregated = genre_counts.groupby('genre')['count'].sum().reset_index()

plt.figure(figsize=(10, 6))
genre_counts_aggregated.sort_values(by='count', ascending=False, inplace=True)
genre_counts_aggregated.head(20).plot(kind='bar', x='genre', y='count')
plt.title("Top 20 Most Popular Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

This is a word cloud of the most popular genres.

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color ='white').generate_from_frequencies(dict(zip(genre_counts_df['genre'], genre_counts_df['count'])))

plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

GENRES ASSOCIATED WITH ARTISTS

This is the distribution of the number of genres each artists has.

```python
csv_files = [
    "data/GENRESPERARTISTS/genresperartists1.csv",
    "data/GENRESPERARTISTS/genresperartists2.csv",
    "data/GENRESPERARTISTS/genresperartists3.csv",
    "data/GENRESPERARTISTS/genresperartists4.csv"
]

df_list = [pd.read_csv(file) for file in csv_files]
genres_per_artist_df = pd.concat(df_list, ignore_index=True)


plt.figure(figsize=(10, 6))

plt.hist(genres_per_artist_df['num_genres'], bins=range(int(genres_per_artist_df['num_genres'].max()) + 1), edgecolor='black')
plt.xlabel('Number of Genres per Artist')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Genres per Artist')
plt.show()
```

This is the same as a boxplot.

```python
plt.figure(figsize=(8, 6))
plt.boxplot(genres_per_artist_df['num_genres'])
plt.xlabel('Number of Genres per Artist')
plt.title('Box Plot of Genres Count per Artist')
plt.show()
```
