```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
genre_counts_df = pd.read_csv("genre_counts.csv")
```

```python
genre_counts_df.plot(kind='bar', x='genre', y='count')
plt.title("Frequency of Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

```python
genre_counts = pd.read_csv("data/genre_counts.csv")


# Aggregate the genre counts
genre_counts_aggregated = genre_counts.groupby('genre')['count'].sum().reset_index()

# Visualization
plt.figure(figsize=(10, 6))
genre_counts_aggregated.sort_values(by='count', ascending=False, inplace=True)
genre_counts_aggregated.head(20).plot(kind='bar', x='genre', y='count')
plt.title("Top 20 Most Popular Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```
