```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

SONG LEVEL AUDIO FEATURES

```python

sns.set(style="whitegrid")

colors = ["#1ED760", "#120D1A", "#82E6A6"]


path_audio = 'SONGLEVEL/'  
all_files_audio = glob.glob(path_audio + "*.csv")
audio_features_df = pd.concat((pd.read_csv(f) for f in all_files_audio))

plt.figure(figsize=(15, 10))
audio_features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'valence']

for i, feature in enumerate(audio_features, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(y=audio_features_df[feature], color=colors[0])
    plt.title(f'Distribution of {feature.capitalize()}')

plt.tight_layout()
plt.show()
```

TRACK DURATION ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob

# Seaborn style
sns.set(style="whitegrid")

# Color palette
colors = ["#1ED760", "#120D1A", "#82E6A6"]

# Read data from a CSV file
path_audio = 'DURATION/'  # Update this path
all_files_audio = glob.glob(path_audio + "*.csv")
audio_features_df = pd.concat((pd.read_csv(f) for f in all_files_audio))

# Create a histogram for duration_ms with color palette
plt.figure(figsize=(8, 6))
sns.histplot(audio_features_df['duration_ms'], bins=70, kde=False, color=colors[0])
plt.xlabel('Duration (ms)')
plt.ylabel('Frequency')
plt.title('Distribution of Song Durations')
plt.show()

```
