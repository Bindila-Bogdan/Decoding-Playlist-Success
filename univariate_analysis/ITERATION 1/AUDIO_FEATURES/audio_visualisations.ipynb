```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

SONG LEVEL AUDIO FEATURES

```python
audio_features_df = pd.read_csv('data/SONGLEVEL/audio_features_summary.csv')

plt.figure(figsize=(15, 10))

audio_features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'valence']

for i, feature in enumerate(audio_features, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(y=audio_features_df[feature])
    plt.title(f'Distribution of {feature.capitalize()}')

plt.tight_layout()
plt.show()
```

We will use a bar plot to visualize the mean values with standard deviation as error bars


```python

audio_features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']

plt.figure(figsize=(14, 8))
# Error bars will represent standard deviation
errors = audio_features_df.loc[audio_features_df['summary'] == 'stddev', audio_features].iloc[0]
means = audio_features_df.loc[audio_features_df['summary'] == 'mean', audio_features].iloc[0]

plt.bar(audio_features, means, yerr=errors, capsize=5)
plt.xlabel('Audio Features')
plt.ylabel('Value')
plt.title('Mean and Standard Deviation of Song Level Audio Features')
plt.show()
```

TRACK DURATION ANALYSIS

```python
duration_stats_df = pd.read_csv("data/TRACKDURATION/track_duration_stats.csv")

plt.figure(figsize=(10, 6))
plt.boxplot(duration_stats_df['duration_ms'])
plt.xlabel('Duration (ms)')
plt.ylabel('Count')
plt.title('Boxplot of Track Duration')
plt.show()
```

MODE DISTRIBUTION

```python
# Load the mode distribution data
mode_df = pd.read_csv("data/MODEDISTRIBUTION/mode_distribution.csv")

# Plot a bar chart for mode distribution
plt.figure(figsize=(8, 6))
plt.bar(mode_df['mode'].astype(str), mode_df['count'])
plt.xlabel('Mode')
plt.ylabel('Count')
plt.title('Distribution of Track Modes')
plt.xticks(rotation=45)  # Ensure the x-ticks are readable
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(mode_df['count'], labels=mode_df['mode'].astype(str), autopct='%1.1f%%')
plt.title('Mode Distribution of Tracks')
plt.show()
```
